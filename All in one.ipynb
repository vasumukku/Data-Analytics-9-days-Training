```python
'''
Two different courses have student lists:
Python class: {"Ravi", "Anita", "Kiran", "Rahul"}
Java class: {"Kiran", "Rahul", "Sneha", "Meena"}

👉 Write a Python program to:
Find students who are learning both Python and Java.
Find students who are learning only Python.
Find students who are learning only Java.
Find all unique students from both classes.
'''

# Student sets
python_class = {"Ravi", "Anita", "Kiran", "Rahul"}
java_class = {"Kiran", "Rahul", "Sneha", "Meena"}



# 1. Students learning both Python and Java
both = python_class.intersection(java_class)

# 2. Students learning only Python
only_python = python_class.difference(java_class)

# 3. Students learning only Java
only_java = java_class.difference(python_class)

# 4. All unique students
all_students = python_class.union(java_class)

# Print results
print("Both Python & Java:", both)
print("Only Python:", only_python)
print("Only Java:", only_java)
print("All Students:", all_students)
```

```python
'''
You and your friend went shopping.
Your list: {"milk", "bread", "eggs", "butter"}
Friend's list: {"bread", "butter", "jam", "cheese"}
👉 Write a Python program to find:
Items both of you bought
Items only you bought
Items only your friend bought
The list of all unique items
'''

# Shopping lists
my_list = {"milk", "bread", "eggs", "butter"}
friend_list = {"bread", "butter", "jam", "cheese"}

# 1. Items both bought
both = my_list.intersection(friend_list)

# 2. Items only I bought
only_me = my_list.difference(friend_list)

# 3. Items only friend bought
only_friend = friend_list.difference(my_list)

# 4. All unique items
all_items = my_list.union(friend_list)

# Print results
print("Both Bought:", both)
print("Only I Bought:", only_me)
print("Only Friend Bought:", only_friend)
print("All Items:", all_items)
```

# NumPy Joining Array

# Joining means combining two or more NumPy arrays into a single array.
# NumPy provides the following functions for joining arrays:

# np.concatenate()
# np.stack()
# np.hstack()
# np.vstack()
# np.dstack()

```python
#Joining Two 1D Arrays using concatenate()

import numpy as np
arr1 = np.array([10, 20, 30])
arr2 = np.array([40, 50, 60])

result = np.concatenate((arr1, arr2))

print(result)
```

```python
#Joining Two 2D Arrays (Row-wise)

import numpy as np

arr1 = np.array([[10, 20],[30, 40]])
arr2 = np.array([[50, 60],[70, 80]])

result = np.concatenate((arr1, arr2), axis=0)

print(result)
```

```python
#Using stack()
import numpy as np

arr1 = np.array([10, 20, 30])
arr2 = np.array([40, 50, 60])

result = np.stack((arr1, arr2))

print(result)
```

```python
#Using hstack() (Horizontal Stack)
import numpy as np

arr1 = np.array([10, 20, 30])
arr2 = np.array([40, 50, 60])

result = np.hstack((arr1, arr2))

print(result)
```

```python
#Using vstack() (Vertical Stack)
import numpy as np

arr1 = np.array([10, 20, 30])
arr2 = np.array([40, 50, 60])

result = np.vstack((arr1, arr2))

print(result)
```

```python
# Using dstack() (Depth Stack)
import numpy as np

arr1 = np.array([10, 20, 30])
arr2 = np.array([40, 50, 60])

result = np.dstack((arr1, arr2))

print(result)
```

# Filtering Data

```python
import pandas as pd

data = {
    "ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Rahul", "Anil", "Priya", "Sneha", "Kiran", "Ravi"],
    "Department": ["BCA", "BCA", "B.Sc", "B.Com", "BCA", "B.Sc"],
    "Marks": [85, 90, 78, 92, 80, 75],
    "City": ["Belagavi", "Hubli", "Dharwad", "Belagavi", "Hubli", "Gadag"]
}

df = pd.DataFrame(data)

print(df)
```

```python
#query()
'''
query() filters rows based on a condition written as a string.
Syntax
df.query("condition")
'''

# Students with Marks Greater Than 80
print(df.query("Marks > 80"))

# Students from BCA Department
print(df.query("Department == 'BCA'"))

#Multiple Conditions
print(df.query("Department == 'BCA' and Marks >= 85"))
```

```python
# isin()
'''
isin() checks whether values exist in a given list.

Syntax
df[column].isin(list)
'''

#Students from Hubli or Gadag
print(df[df["City"].isin(["Hubli", "Gadag"])])

#Students from BCA or B.Sc
print(df[df["Department"].isin(["BCA", "B.Sc"])])

```

```python
#between()
'''
between() selects values that lie within a specified range.

Syntax
df[column].between(start, end)
'''
#Marks Between 80 and 90
print(df[df["Marks"].between(80, 90)])

# ID Between 102 and 105
print(df[df["ID"].between(102, 105)])

```

```python
import pandas as pd

# Create DataFrame
employee = {
    "Emp_ID": [101,102,103,104,105],
    "Name": ["Amit","Neha","Ravi","Pooja","Karan"],
    "Department": ["HR","IT","Finance","IT","HR"],
    "Salary": [25000,40000,35000,50000,28000]
}

df = pd.DataFrame(employee)

print("Original DataFrame")
print(df)

# ----------------------------------------------------
# 1. Insert a new column Age after the Name column
# ----------------------------------------------------
df.insert(2, "Age", [25, 30, 28, 35, 27])

print("\n1. Age Column Added")
print(df)

# ----------------------------------------------------
# 2. Insert a new column City at the last position
# ----------------------------------------------------
df.insert(len(df.columns), "City",
          ["Belagavi", "Hubli", "Dharwad", "Bengaluru", "Mysuru"])

print("\n2. City Column Added")
print(df)

# ----------------------------------------------------
# 3. Remove the Department column
# ----------------------------------------------------
df = df.drop("Department", axis=1)

print("\n3. Department Column Removed")
print(df)

# ----------------------------------------------------
# 4. Remove the employee with index 2
# ----------------------------------------------------
df = df.drop(2)

print("\n4. Employee at Index 2 Removed")
print(df)

# ----------------------------------------------------
# 5. Rename Salary to Monthly_Salary
# ----------------------------------------------------
df = df.rename(columns={"Salary": "Monthly_Salary"})

print("\n5. Salary Column Renamed")
print(df)

# ----------------------------------------------------
# 6. Rename Name to Employee_Name
# ----------------------------------------------------
df = df.rename(columns={"Name": "Employee_Name"})

print("\n6. Name Column Renamed")
print(df)

# ----------------------------------------------------
# 7. Add Bonus Column
# ----------------------------------------------------
df = df.assign(Bonus=[2000,3000,4000,2200])

print("\n7. Bonus Column Added")
print(df)

# ----------------------------------------------------
# 8. Create Total_Salary Column
# ----------------------------------------------------
df = df.assign(
    Total_Salary=df["Monthly_Salary"] + df["Bonus"]
)

print("\n8. Total Salary Column Added")
print(df)

# ----------------------------------------------------
# 9. Create Tax Column (10% of Salary)
# ----------------------------------------------------
df = df.assign(
    Tax=df["Monthly_Salary"] * 0.10
)

print("\n9. Tax Column Added")
print(df)

# ----------------------------------------------------
# 10. Final DataFrame
# ----------------------------------------------------
print("\n10. Final DataFrame")
print(df)
```

```python
import pandas as pd
import numpy as np

# Create DataFrame
employee = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Neha", np.nan, "Pooja", "Karan", "Ravi"],
    "Department": ["HR", "IT", "Finance", np.nan, "HR", "Sales"],
    "Salary": [25000, np.nan, 35000, 50000, 28000, np.nan]
}

df = pd.DataFrame(employee)

print("=" * 60)
print("ORIGINAL DATAFRAME")
print("=" * 60)
print(df)

# ----------------------------------------------------
# 1. Display missing values
# ----------------------------------------------------
print("\n1. Missing Values (isnull())")
print(df.isnull())

# ----------------------------------------------------
# 2. Count missing values
# ----------------------------------------------------
print("\n2. Count Missing Values")
print(df.isnull().sum())

# ----------------------------------------------------
# 3. Display non-missing values
# ----------------------------------------------------
print("\n3. Non-Missing Values (notnull())")
print(df.notnull())

# ----------------------------------------------------
# 4. Replace missing Name with 'Unknown'
# ----------------------------------------------------
df1 = df.copy()
df1["Name"] = df1["Name"].fillna("Unknown")

print("\n4. Replace Missing Name with 'Unknown'")
print(df1)

# ----------------------------------------------------
# 5. Replace missing Salary with 0
# ----------------------------------------------------
df2 = df.copy()
df2["Salary"] = df2["Salary"].fillna(0)

print("\n5. Replace Missing Salary with 0")
print(df2)

# ----------------------------------------------------
# 6. Replace missing Salary with Average Salary
# ----------------------------------------------------
df3 = df.copy()
df3["Salary"] = df3["Salary"].fillna(df3["Salary"].mean())

print("\n6. Replace Missing Salary with Average Salary")
print(df3)

# ----------------------------------------------------
# 7. Replace missing Department with 'Not Assigned'
# ----------------------------------------------------
df4 = df.copy()
df4["Department"] = df4["Department"].fillna("Not Assigned")

print("\n7. Replace Missing Department")
print(df4)

# ----------------------------------------------------
# 8. Remove rows with missing values
# ----------------------------------------------------
print("\n8. Remove Rows with Missing Values")
print(df.dropna())

# ----------------------------------------------------
# 9. Remove columns with missing values
# ----------------------------------------------------
print("\n9. Remove Columns with Missing Values")
print(df.dropna(axis=1))

# ----------------------------------------------------
# 10. Remove rows where Department is missing
# ----------------------------------------------------
print("\n10. Remove Rows with Missing Department")
print(df.dropna(subset=["Department"]))

print("\n" + "=" * 60)
print("PROGRAM COMPLETED SUCCESSFULLY")
print("=" * 60)
```

# Value Analysis
# Value Analysis methods are used to analyze the values in a column. They help us find:

# How many times each value appears
# The unique values in a column
# The total number of unique values

```python
# value_counts()
# Returns the number of times each unique value appears in a column.
'''
Syntax
df["Column_Name"].value_counts()
'''

# Count Employees in Each Department
print(df["Department"].value_counts())

# Count Employees in Each City
print(df["City"].value_counts())

#===============================================
# unique()
Returns all unique values from a column.
'''
Syntax
df["Column_Name"].unique()
'''
#exa1
print(df["Department"].unique())

#exa2
print(df["City"].unique())

#=================================================
# nunique()
# Returns the total number of unique values in a column.
'''
Syntax
df["Column_Name"].nunique()
'''
#exa1
print(df["Department"].nunique())

#exa2
print(df["Department"].nunique())

#exa3
print(df["City"].nunique())

```

# Grouping Data
# The groupby() method is used to group rows that have the same value in one or more columns and then
# perform operations such as sum, mean, count, max, min, etc.

# It is commonly used to analyze grouped data, such as:

# Total salary by department
# Average marks by class
# Total sales by city
# Number of employees in each department

```python
# Syntax
df.groupby("Column_Name")
```

```python
# Group by Department
print(df.groupby("Department"))
```

```python
# Total Salary by Department
print(df.groupby("Department")["Salary"].sum())
```

```python
# Average Salary by Department
print(df.groupby("Department")["Salary"].mean())
```

```python
# Highest Salary by Department
print(df.groupby("Department")["Salary"].max())
```

```python
# Lowest Salary by Department
print(df.groupby("Department")["Salary"].min())
```

```python
# Employee Count by Department
print(df.groupby("Department")["Emp_ID"].count())
```

```python
# Group by City
print(df.groupby("City")["Salary"].sum())
```

```python
# Group by Multiple Columns
# Group by Department and City.

print(df.groupby(["Department", "City"])["Salary"].sum())
```

```python
# Multiple Calculations
print(df.groupby("Department")["Salary"].agg(["sum", "mean", "max", "min", "count"]))
```

# Combining DataFrames
# Sometimes data is stored in multiple DataFrames. Pandas provides methods to combine them into a single DataFrame.

```python
import pandas as pd

df1 = pd.DataFrame({
    "Emp_ID": [101, 102, 103],
    "Name": ["Amit", "Neha", "Ravi"]
})

df2 = pd.DataFrame({
    "Emp_ID": [104, 105, 106],
    "Name": ["Pooja", "Karan", "Meena"]
})

print("DataFrame 1")
print(df1)

print("\nDataFrame 2")
print(df2)
```

```python
#concat()
#Combines two or more DataFrames.
'''
Syntax
pd.concat([df1, df2])
'''

# Example 1: Vertical Concatenation
result = pd.concat([df1, df2])
print(result)


#Example 2: Horizontal Concatenation
df3 = pd.DataFrame({"Department": ["HR", "IT", "Finance"]})
result = pd.concat([df1, df3], axis=1)
print(result)


# merge()
'''
Combines DataFrames using a common column.
'''
# Sample Data
employee = pd.DataFrame({
    "Emp_ID": [101,102,103],
    "Name": ["Amit","Neha","Ravi"]
})

salary = pd.DataFrame({
    "Emp_ID": [101,102,103],
    "Salary": [25000,40000,35000]
})

#example
result = pd.merge(employee, salary, on="Emp_ID")
print(result)


# join()
'''
Joins DataFrames using their indexes.
'''
# Example
employee = pd.DataFrame({"Name": ["Amit", "Neha", "Ravi"]})
salary = pd.DataFrame({"Salary": [25000, 40000, 35000]})
result = employee.join(salary)
print(result)

```

# Saving Data
# After processing the data, we can save it into different file formats.

```python
import pandas as pd

employee = {
    "Emp_ID": [101,102,103],
    "Name": ["Amit","Neha","Ravi"],
    "Department": ["HR","IT","Finance"],
    "Salary": [25000,40000,35000]
}

df = pd.DataFrame(employee)
```

```python
# to_csv()
# Saves the DataFrame as a CSV file.
'''
Syntax
df.to_csv("employee.csv")
'''

#Example
df.to_csv("employee.csv", index=False)
print("CSV File Saved Successfully")

# to_excel()
#Saves the DataFrame as an Excel file.
'''
Syntax
df.to_excel("employee.xlsx")
'''
# Example
df.to_excel("employee.xlsx", index=False)
print("Excel File Saved Successfully")
```
