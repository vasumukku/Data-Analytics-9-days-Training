```python
import matplotlib.pyplot as plt

x=season_avg.index
y=season_avg.values
plt.plot(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt

season_avg = df.groupby("season")["runs"].mean()

plt.figure(figsize=(10,5))

plt.plot(
    season_avg.index,
    season_avg.values,
    marker="o",
    color="blue",
    linewidth=3
)

plt.title("Average Runs Season Wise")
plt.xlabel("Season")
plt.ylabel("Average Runs")

plt.grid(True)

plt.show()
```

```python
df["strikeRate"] = pd.to_numeric(
    df["strikeRate"],
    errors="coerce"
)
```

```python
strike = df.groupby("season")["strikeRate"].mean()
```

```python
plt.figure(figsize=(10,5))

plt.plot(
    strike.index,
    strike.values,
    marker="D",
    color="purple",
    linewidth=3
)

plt.title("Average Strike Rate Season Wise")
plt.xlabel("Season")
plt.ylabel("Strike Rate")

plt.grid()

plt.show()
```

## Bar plot

```python
import matplotlib.pyplot as plt

top_players = (
    df.groupby("name")["runs"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(12,6))

plt.bar(
    top_players.index,
    top_players.values,
    color="royalblue",
    edgecolor="black"
)

plt.title("Top 10 Run Scorers")
plt.xlabel("Players")
plt.ylabel("Total Runs")

plt.xticks(rotation=45)

plt.grid(axis="y")

plt.show()
```

## piechart

```python
import matplotlib.pyplot as plt

home_team = df["home_team"].value_counts()

plt.figure(figsize=(8,8))

plt.pie(
    home_team.values,
    labels=home_team.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Home Team Percentage")

plt.show()
```

```python
venue = df["venue"].value_counts().head(8)

plt.figure(figsize=(8,8))

plt.pie(
    venue.values,
    labels=venue.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Top 8 Venue Percentage")

plt.show()
```

```python
country = df["country"].value_counts()

plt.figure(figsize=(8,8))

plt.pie(
    country.values,
    labels=country.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Country Percentage")

plt.show()
```
